#16. 將文件分為N份

以命令行參數等形式接收自然數N
，按「行」為單位將原檔分割成N
份。 使用命令確認結果。

```python
#N =2 or 3
#會用到迴圈
import math
N = 3
with open('popular-names.txt', 'r') as f:
    lines = f.readlines()
total_lines = len(lines)
lines_per_file = math.ceil(total_lines / N)
print(f"總行數: {total_lines}")
print(f"每份約: {lines_per_file} 行")
for i in range(N):
    start = i * lines_per_file
    end = start + lines_per_file
    chunk = lines[start:end]
    with open(f'part_{i+1}.txt', 'w') as f:
        f.writelines(chunk)
    print(f"part_{i+1}.txt 建立完成，共 {len(chunk)} 行")
```
